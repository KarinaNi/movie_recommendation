```python
import numpy as np
import pandas as pd
import os
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(2)
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on = 'title')
movies.head(2)
```

```python
movies = movies[['movie_id','title','overview','genres','cast','crew']]
movies.dropna(inplace=True)
```

```python
type(movies['genres'])
```

```python
#trying to get them in a list
import json
import ast
def dict_to_list(material):
    return [item['name'] for item in ast.literal_eval(material)]
```

```python
movies['genres'] = movies['genres'].apply(dict_to_list)
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tv = TfidfVectorizer()
cv = CountVectorizer()
```

```python
tvector = tv.fit_transform(movies['title']).toarray()
cvector = cv.fit_transform(movies['title']).toarray()
```

```python
tsimilar = cosine_similarity(tvector)
csimilar = cosine_similarity(cvector)
```

```python
tsimilar
```

```python
def trecommend(title):
    index = movies[movies['title'] == title].index[0]
    distances = sorted(list(enumerate(tsimilar[index])), reverse=True, key=lambda x: x[1])
    for rec in distances[1:10]:
        print(movies.iloc[rec[0]].title)
```

```python
trecommend('Star Wars')
```

```python
def crecommend(title):
    index = movies[movies['title'] == title].index[0]
    distances = sorted(list(enumerate(csimilar[index])), reverse=True, key=lambda x: x[1])
    for rec in distances[1:10]:
        print(movies.iloc[rec[0]].title)
```

```python
crecommend('Star Wars')
```
